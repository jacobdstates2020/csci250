# Objective
* Learn the basics of the CSCI AutoGrader.

# What is the AutoGrader?

A tool created by Mines CS faculty to test programming exercises. 

Used for quizzes to check completion of the software topics.

We will import the AutoGrader scores to Canvas - it is not automatic.

* **Use VPN** to connect to the Mines network from off campus. 
* Read about **Virtual Private Networks** on the [ITS site](https://its.mines.edu/software-title/vpn/).
</div>



<div class="alert alert-block alert-warning">

</div>



<div class="alert alert-block alert-warning">

# Step 3: select the relevant quiz
</div>
    


<div class="alert alert-block alert-warning">
    
# Step 4: select a problem
* Each quiz has a set of problems to be completed for credit.
* You can answer the questions in any order you like.
</div>




<div class="alert alert-block alert-warning">
    
# Step 5: solve the problem

For each problem:
* read the instructions on the left 
* write your answer on the right
* click **SUBMIT** to test your answer
* multiple submission are allowed
</div>

<div class="alert alert-block alert-warning">
    
# Step 6: check your scores

* select **Scores** from the top bar, next to **Preferences**
* select CSCI 250 from the list, and then the quiz
* AutoGrader records the highest score of all submissions
    
</div>
