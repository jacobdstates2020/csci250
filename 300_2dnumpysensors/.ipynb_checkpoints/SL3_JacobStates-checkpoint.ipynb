# Data noise

Data provided by physical sensors are often corrupted by noise. 

Two common types of noise are 
* **random**, i.e. continuous small fluctuations, and
* **spiky**, i.e. occasional large fluctuations.

# Data denoising

We can denoise data using digital filtering:
* **mean filters** - replace a value of the time series with an average of the neighbor values within a specified window. 

* **median filters** - replace a value of the time series with the median of the neighbor values within a specified window. 

Consider the time series defined below representing a damped oscillatory signal, 

$$
a(t) = e^{-kt} \sin(2 \pi f t)
$$

corrupted by uniform random noise and lots of large spikes. 

Your task is to filter the data using mean and median filters (use a window equal to 1/4 of the signal period).

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

import numpy.random as rng
rng.seed(1001)
```

```python
T = 10                     # [s]  total time
F = 0.01                   # [Hz] sampling rate

nt = int(T/F)              # samples
f  = 0.5                   # signal frequency
k  = 0.2                   # signal damping
p = 1/f                  # signal period

t = np.linspace(0,10,nt)   # time axis
```

```python
s = np.sin(p*pi*f*t) * np.exp(-k*t)

# start with the signal
a = s.copy()

# add white noise
wn = rng.uniform(-0.2,+0.2, nt)
a += wn

# add spiky noise
ns = 51
sn = np.zeros(nt)
sn[ rng.randint(1,nt,ns) ] = rng.uniform(-5.0,+5.0,ns)
a += sn
```

```python
plt.figure(figsize=(15,5))

plt.title('Original Noisey Signal')
plt.plot(t,a,'k')
plt.xlabel('Time') 
plt.ylabel('Amplitude') 
plt.ylim(-5.1,+5.1);
```

```python
import statistics as stat

# Moving average filter
def moving_avg(input, win):
    avg = []
    mov_sum = sum(input[:int(win)])
    avg.append(mov_sum/win)
    for i in range(int(len(input) - win)):
        mov_sum += (input[int(i + win)] - input[i]) # Shifts window to the right by one unit
        avg.append(mov_sum/win)
        
    return avg

# Moving median filter
def median(input, win):
    med = []
    mov_med = np.median(input[:int(win)])
    med.append(mov_med)
    for i in range(int(len(input) - win)):
        mov_med += (input[int(i + win)] - input[i])
        med.append(mov_med)
        
    return med

w = nt*p/(T*4)
f = a.copy()

fun = moving_avg(f, w)
fun2 = median(f, w)
```

```python
# Mean Filter Graph
l = np.linspace(0, 10, len(fun))

plt.figure(figsize=(15,5))
plt.plot(t,a,'k', label='Original Signal', linestyle='dashed')
plt.plot(l,fun, color='b', label='Mean Filter')

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Mean Filter applied to signal')
plt.legend()
```

```python
# Median Filter Graph
l2 = np.linspace(0, 10, len(fun2))

plt.figure(figsize=(15,5))
plt.plot(t,a,'k', label='Original Signal',linestyle='dashed')
plt.plot(l2,fun2, label='Median Filter', color='y')

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Median Filter applied to signal')
plt.legend()
```

The filters are not great. The mean filter works alright. The median filter ruins the original signal amplitude but preserves the shape. For both filters I coded a moving average and median with a window with length of 50 samples or 1/4 the length of the original period of 200 samples. It looks like when period increases, the filters effectiveness decreases. When the window increases the mean filter is more effective but the median filter becomes smoother but more inaccurate.

I could not locate the scipy functions for a median filter analog. I did however find the convolve function which was the closest analog to the mean filter. The convolve is much smoother with a window of length 50 than the mean filter. I'm curious if I did something wrong or not.

```python
# Convolution Filter
import scipy.signal as signal

win = signal.windows.hann(50)
fun3 = signal.convolve(f, win)/sum(win)

l = np.linspace(0, 10, len(fun3))

plt.figure(figsize=(15,5))
plt.plot(t,a,'k', label='Original Signal', linestyle='dashed')
plt.plot(l,fun3, color='b', label='Scipy Convolve Filter')

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Scipy Convolve Filter applied to signal')
plt.legend()
```
