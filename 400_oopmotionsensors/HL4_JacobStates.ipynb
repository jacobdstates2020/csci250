## class `Accelerometer`

Required methods:
* `__init__(tref=0)`
    * set instance variable `tref` 
        * if the constructor argument is not 0, set `self.tref=tref`
        * otherwise, set `self.tref` to the current time
    * initializes 4 `numpy` arrays for time and the 3 acceleration components
        * set all the array sizes to 0
        * set all the array values to 0

* `getTref()`
    * returns the reference time `tref`
* `setTref(tref)`
    * set the time reference to `tref`

* `read()`
    * reads one sample (ax,ay,az) from the accelerometer with `accUtil.py`
    * appends the relative time to `tref` to the declared `numpy` array
    * appends the acceleration components to the declared `numpy` arrays
        
* `save(name=None)`
    * saves the time and accelerometer data to an `.npz` file
        
* `plot()`
    * create and display a single plot of the acceleration data as a function of time
    * add labels, legend, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
import RPi.GPIO as GPIO
GPIO.setmode(GPIO.BCM)
import time
import adcUtil as adc
import accUtil as acc


class Accelerometer:
    '''class Accelerometer describes spacial and time numpy data arrays collected from accelerometer'''
    
    def __init__(self, tref=0):              # tref=0 allows us to not give a value when initializing
        if tref != 0 :
            self.tref = tref                 # Sets tref to what I choose
        else:
            self.tref = time.time()          # Sets tref to current time when initialized if no input since defaults tref=0 
        self.at = np.zeros(0, dtype='float') # Creates 5 numpy arrays that has dim 0 and values 0
        self.ax = np.zeros(0, dtype='float') 
        self.ay = np.zeros(0, dtype='float')
        self.az = np.zeros(0, dtype='float')
        self.ar = np.zeros(0, dtype='float') # Acceleration magnitude array
    
    def getTref(self):
        '''displays reference time tref'''
        print('tref is ',self.tref)
        
    def setTref(self, tref):
        '''sets reference time, tref, to input value'''
        self.tref = tref
        print('tref set to ', self.tref)
        
    def read(self):
        acc_x, acc_y, acc_z = acc.readACC()    # Reads accelerometer and assigns values
        t = time.time() - self.tref            # Sets time read, t, relative to the tref
        ar = np.sqrt((acc_x)*(acc_x) + (acc_y)*(acc_y) + (acc_z)*(acc_z)) # Calculates magnitude
        self.at = np.append(self.at, t)        # Appends t to time array
        self.ax = np.append(self.ax, acc_x)    # Appends accelerometer data to respective arrays
        self.ay = np.append(self.ay, acc_y)
        self.az = np.append(self.az, acc_z)
        self.ar = np.append(self.ar, ar)
        return self.at, self.ax, self.ay, self.az, self.ar # Returns arrays
    
    def save(self, name=None):
        if name != None:
            self.name = name
        else:
            print('Enter file name for acc data:')
            self.name = input()
        np.savez(self.name +'.npz', self.at, self.ax, self.ay, self.az, self.ar)
        print('File saved as:',self.name +'.npz')
    
    def plot(self):
        plt.figure(figsize=(10,5))
        plt_acc = plt.plot(self.at, self.ar, label='Absolute Acceleration')
        plt.xlabel('Time (s)')
        plt.ylabel('Absolute Acceleration (m/s^2)')
        plt.legend()
        plt.title('Accelerometer Data')
        plt.show()
    
    def compplot(self):
        plt.figure(figsize=(10,5))
        plt_acc = plt.plot(self.at, self.ax, label='x Acceleration')
        plt_acc = plt.plot(self.at, self.ay, label='y Acceleration')
        plt_acc = plt.plot(self.at, self.az, label='z Acceleration')
        plt.xlabel('Time (s)')
        plt.ylabel('Component Acceleration (m/s^2)')
        plt.legend()
        plt.title('Accelerometer Data')
        plt.show()
```

## class Piezoelement

Required methods:
* `__init__(tref=0, chan=0)`
    * set instance variable `tref` 
        * if the constructor argument is not 0, set `self.tref=tref`
        * otherwise, set `self.tref` to the current time
    * set the instance variable `chan`
    * initializes 2 `numpy` arrays for time and the piezoelement data
        * set all the array sizes to 0
        * set all the array values to 0

* `getTref()`
    * returns the reference time `tref`       
* `setTref(tref)`
    * set the time reference to `tref`
    
* `read()`
    * reads one sample from the piezoelment with `adcUtil.py`
    * appends the relative time to `tref` to the declared `numpy` array
    * appends the piezoelement data to the declared `numpy` array
    
* `save(name=None)`
    * saves the time and piezoelement data to an `.npz` file
    
* `plot()`
    * create and display a single plot of the piezoelement data as a function of time
    * add labels, legend, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
import RPi.GPIO as GPIO
GPIO.setmode(GPIO.BCM)
import time
import adcUtil as adc
import accUtil as acc

class Piezoelement:
    '''class Piezoelement describes data collected from piezoelement'''
    
    def __init__(self, tref=0, chan=0):
        if tref != 0 :
            self.tref = tref                 # Sets tref to what I choose
        else:
            self.tref = time.time()
        if chan != 0 :
            self.chan = chan                
        else:
            self.chan = 0
        self.pt = np.zeros(0, dtype='float') # Array for piezo time 
        self.pv = np.zeros(0, dtype='float') # Array for piezo voltage
        
    def getTref(self):
        '''displays reference time tref'''
        print('tref is ',self.tref)
        
    def setTref(self, tref):
        '''sets reference time, tref, to input value'''
        self.tref = tref
        print('tref set to ', self.tref)
    
    def read(self):
        pv = adc.readADC(self.chan, device=0)
        t = time.time() - self.tref
        self.pt = np.append(self.pt, t)
        self.pv = np.append(self.pv, pv)
        return self.pt, self.pv
    
    def save(self, name=None):
        if name != None:
            self.name = name
        else:
            print('Enter file name for piezo data:')
            self.name = input()
        np.savez(self.name +'.npz', self.pt, self.pv)
        print('File saved as:',self.name +'.npz')

    def plot(self):
        plt.figure(figsize=(10,5))
        plt_acc = plt.plot(self.pt, self.pv, label='Piezoelement Voltage')
        plt.xlabel('Time (s)')
        plt.ylabel('Voltage (V)')
        plt.legend()
        plt.title('Piezoelement Voltage Data')
        plt.show()
```
